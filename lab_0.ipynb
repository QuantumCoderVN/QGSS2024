B√†i t·∫≠p 2.5 Vi·∫øt m·ªôt h√†m ki·ªÉm tra m·ªôt ma tr·∫≠n c√≥ ph·∫£i l√† ma tr·∫≠n Hermitian ho·∫∑c Unita hay kh√¥ng.

```python
import numpy as np
def is_hermitian (matrix):
    return np.allclose(matrix, np.conjugate(np.transpose(matrix)))
def is_unitary (matrix):
    identity = np.eye(np.size(matrix, 0))
    return np.allclose(np.dot(matrix, np.conjugate(np.transpose(matrix))), identity)
# V√≠ d·ª•
A = [
    [1+0j, 2-1j],
    [2+1j, 3+0j]
]

U = [
    [1/(2**0.5), 1/(2**0.5)],
    [1/(2**0.5), -1/(2**0.5)]
]

print("Matrix A is Hermitian:", is_hermitian(A))
print("Matrix U is Unitary:", is_unitary(U))
```

# Sử dụng qiskit để lập trình các trạng thái lượng tử GHZ và W

## Trạng thái GHZ 

Điều này có nghĩa là hệ thống sẽ ở trạng thái vướng víu giữa hai trạng thái 
∣000⟩ và ∣111⟩. Nếu đo tất cả các qubit trong hệ thống này, chúng sẽ luôn cho cùng một kết quả: hoặc tất cả đều là 0, hoặc tất cả đều là 1.

### Tính Chất
Vướng víu cực đại: Tất cả các qubit trong trạng thái GHZ đều vướng víu với nhau một cách cực đại. Bất kỳ phép đo nào trên một qubit cũng sẽ ảnh hưởng ngay lập tức đến trạng thái của các qubit còn lại.
Ứng dụng: Trạng thái GHZ được sử dụng trong nhiều ứng dụng lượng tử như mật mã lượng tử, kiểm tra tính phi cục bộ của cơ học lượng tử, và các giao thức trao đổi khóa lượng tử.
### Đo Lường Trạng Thái GHZ
Khi đo các qubit trong trạng thái GHZ, kết quả đo sẽ luôn là một trong hai trạng thái 
∣00...0⟩ hoặc 
∣11...1⟩. Xác suất của mỗi trạng thái sẽ là 50% nếu không có sự can thiệp từ môi trường bên ngoài.

Trạng thái GHZ minh chứng cho tính phi cục bộ của cơ học lượng tử và là một công cụ quan trọng trong nghiên cứu và ứng dụng lượng tử hiện đại.

### Tính Phi Cục Bộ của Cơ Học Lượng Tử
#### Định Nghĩa
Tính phi cục bộ trong cơ học lượng tử đề cập đến hiện tượng mà hai hoặc nhiều hệ lượng tử có thể vướng víu với nhau theo cách mà trạng thái của chúng không thể mô tả một cách độc lập. Các phép đo trên một hệ sẽ ảnh hưởng ngay lập tức đến trạng thái của các hệ khác, bất kể khoảng cách không gian giữa chúng.

#### Lý Thuyết
EPR Paradox: Năm 1935, Einstein, Podolsky, và Rosen (EPR) đã đưa ra một bài báo nổi tiếng, trong đó họ cho rằng cơ học lượng tử không hoàn chỉnh vì nó cho phép các tác động "ma quái" ở khoảng cách xa. Họ đề xuất rằng cần có các biến ẩn địa phương để giải thích hiện tượng này.
Bell's Theorem: Năm 1964, John Bell đã phát triển một định lý chứng minh rằng không có lý thuyết biến ẩn địa phương nào có thể tái tạo tất cả các dự đoán của cơ học lượng tử. Bell's Theorem cho thấy rằng nếu các biến ẩn tồn tại, chúng phải có tính phi cục bộ.

#### Ứng Dụng
- Mật mã lượng tử: Tính phi cục bộ được sử dụng trong các giao thức mật mã lượng tử để đảm bảo tính bảo mật của việc trao đổi thông tin.
- Thông tin lượng tử: Các thuật toán lượng tử và các giao thức truyền thông lượng tử, chẳng hạn như dịch chuyển lượng tử, dựa vào tính phi cục bộ để hoạt động.
#### Ví Dụ Minh Họa: Trạng Thái GHZ
Trạng thái GHZ là một ví dụ cụ thể minh họa tính phi cục bộ. Với trạng thái GHZ, nếu bạn đo một trong các qubit và nhận được kết quả, bạn ngay lập tức biết trạng thái của các qubit còn lại mà không cần phải đo trực tiếp chúng. Điều này xảy ra ngay cả khi các qubit ở khoảng cách xa nhau, minh chứng cho tính phi cục bộ của cơ học lượng tử.

```python
# from qiskit.quantum_info import Statevetor 
from qiskit.visualization import plot_bloch_multivector 
from qiskit import QuantumCircuit 
# Tạo mạch lượng tử với 3 qubit và 3 qubit đo
qc = QuantumCircuit (3, 3)

# Tạo trạng thái GHZ
qc.h(0)
qc.cx(0, 1)
qc.cx(0, 2)

# Hiển thị mạch 
qc.draw('mpl')
```

```python
# Hiển thị trạng thái của các qubit trên quả cầu Bloch
plot_bloch_multivector(qc)
```

```python
from qiskit.providers.basic_provider import BasicSimulator 
from qiskit.visualization import plot_histogram
# Thực hiện đo các qubit 
qc.measure([0, 1, 2], [0, 1, 2])

backend = BasicSimulator ()
result = backend.run(qc, shots = 1000).result()
counts = result.get_counts()
plot_histogram(counts)
```

## Trạng thái W

Trạng Thái W Cho Ba Qubit
Trạng thái W là một trạng thái lượng tử vướng víu của ba qubit, có dạng trong ký hiệu bra-ket như sau:



- Trạng Thái W Tổng Quát Cho n Qubit
Trạng thái W cho 𝑛 qubit, tổng quát hóa từ trạng thái W cho ba qubit, là một siêu vị trí lượng tử trong đó đúng một qubit ở trạng thái kích thích ∣1⟩ và tất cả các qubit còn lại ở trạng thái cơ bản ∣0⟩. 
- Trạng thái này được viết như sau: 




- Đặc Điểm của Trạng Thái W
    - Vướng Víu Đa Hạt: Trạng thái W đại diện cho một loại vướng víu đa hạt đặc biệt, không thể biến đổi thành trạng thái GHZ và ngược lại bằng các phép toán lượng tử cục bộ (LOCC).
    - Khả Năng Chống Lại Mất Qubit: Nếu một qubit trong trạng thái W bị mất, trạng thái của các qubit còn lại vẫn duy trì vướng víu. Điều này làm cho trạng thái W khác biệt so với trạng thái GHZ, vốn trở nên hoàn toàn phân ly khi mất một qubit.
    - Ứng Dụng Trong Lưu Trữ Lượng Tử: Tính chất chống lại mất qubit làm cho trạng thái W hữu ích trong các ứng dụng lưu trữ lượng tử, nơi sự mất mát hạt không ảnh hưởng nghiêm trọng đến tính toàn vẹn của dữ liệu.
- Ứng Dụng của Trạng Thái W
    - Thông Tin Lượng Tử: Trạng thái W được sử dụng trong nhiều giao thức và thuật toán thông tin lượng tử.
    - Mật Mã Lượng Tử: Trạng thái W có ứng dụng trong mật mã lượng tử nhờ vào tính vướng víu và khả năng bảo vệ thông tin khỏi sự mất mát qubit.
    - Lưu Trữ Lượng Tử: Trong các hệ thống mà một qubit được lưu trữ trong một tập hợp nhiều hệ thống hai mức, "1" logic thường được biểu diễn bằng trạng thái W, trong khi "0" logic được biểu diễn bằng trạng thái ∣00...0⟩.
    - Trạng thái W là một công cụ mạnh mẽ trong lĩnh vực thông tin lượng tử nhờ vào tính vướng víu đa hạt đặc biệt và khả năng chống lại mất qubit, giúp bảo vệ và duy trì thông tin lượng tử trong nhiều ứng dụng khác nhau.

```python
# Tạo mạch lượng tử với 3 qubit và 3 bit do 
qc = QuantumCircuit(3, 3)

# Tạo trạng thái W
qc.ry(2*1.23096, 0)
qc.ch (0, 1)
qc.cx (1, 2)
qc.cx (0, 1)
qc.x(0)

# Hiển thị mạch 
qc.draw('mpl')
```

```python
# Hiển thị trạng thái của các qubit trên quả cầu Bloch
plot_bloch_multivector(qc)
```

```python
# Thực hiện đo các qubit 
qc.measure([0, 1, 2], [0, 1, 2])

backend = BasicSimulator ()
result = backend.run(qc, shots = 1000).result()
counts = result.get_counts()
plot_histogram(counts)
```

### B√†i t·∫≠p 3.1 Vi·∫øt m·ªôt ch∆∞∆°ng tr√¨nh m√¥ ph·ªèng l·∫°i th√≠ nghi·ªám v·ªÅ nh·ªØng vi√™n bi ƒë∆∞·ª£c m√¥ t·∫£ t·ª´ ph·∫ßn 3.1. Ch∆∞∆°ng tr√¨nh c·∫ßn nh·∫≠p v√†o ma tr·∫≠n Boolean, tr·∫°ng th√°i kh·ªüi t·∫°o (s·ªë vi√™n bi tr√™n m·ªói ƒë·ªânh) v√† s·ªë l·∫ßn ho·∫°t ƒë·ªông ùëõ. K·∫øt qu·∫£ tr·∫£ v·ªÅ l√† tr·∫°ng th√°i h·ªá th·ªëng sau ùëõ ho·∫°t ƒë·ªông.

```python
import numpy as np

def simulate_experiment(matrix, initial_state, n):
    # Chuy·ªÉn ma tr·∫≠n v√† tr·∫°ng th√°i kh·ªüi t·∫°o th√†nh m·∫£ng numpy
    matrix = np.array(matrix, dtype=bool).astype(int)
    state = np.array(initial_state, dtype=int).reshape(-1, 1)
    
    # Nh√¢n ma tr·∫≠n v·ªõi vector tr·∫°ng th√°i ban ƒë·∫ßu
    for _ in range(n):
        state = np.dot(matrix, state)
    
    return state

#V√≠ d·ª• 
boolean_matrix = [
    [False, False, False, False, False, False],
    [False, False, False, False, False, False],
    [False, True, False, False, False, True],
    [False, False, False, True, False, False],
    [False, False, True, False, False, False],
    [True, False, False, False, True, False],
]

initial_state = [6, 2, 1, 5, 3, 10]
n = 1
result = simulate_experiment(boolean_matrix, initial_state,n)
print (np.array(initial_state).reshape(-1, 1)) 
print(result)  
```

### B√†i t·∫≠p 3.2 N√¢ng c·∫•p b√†i t·∫≠p 3.1 v·ªõi y√™u c·∫ßu ph·∫ßn t·ª≠ c·ªßa ma tr·∫≠n c√≥ th·ªÉ l√† ph√¢n s·ªë.

```python
import numpy as np
from fractions import Fraction

def simulate_fraction_experiment(matrix, initial_state, n):
    # Chuy·ªÉn ma tr·∫≠n v√† tr·∫°ng th√°i kh·ªüi t·∫°o th√†nh m·∫£ng numpy v·ªõi ki·ªÉu Fraction
    matrix = np.array(matrix, dtype=object)
    state = np.array(initial_state, dtype=object).reshape(-1, 1)
    
    # L·∫∑p l·∫°i qu√° tr√¨nh c·∫≠p nh·∫≠t tr·∫°ng th√°i n l·∫ßn
    for _ in range(n):
        state = np.dot(matrix, state)
    
    return state

# ƒê·∫ßu v√†o c·ª• th·ªÉ c·ªßa b√†i to√°n v·ªõi ph·∫ßn t·ª≠ l√† ph√¢n s·ªë
fraction_matrix = [
    [Fraction(0, 1), Fraction(1, 6), Fraction(5, 6)],
    [Fraction(1, 3), Fraction(1, 2), Fraction(1, 6)],
    [Fraction(2, 3), Fraction(1, 3), Fraction(0, 1)],
]

initial_state = [Fraction(1, 6), Fraction(1, 6), Fraction(2, 3)]
iterations = 1  # S·ªë l·∫ßn ho·∫°t ƒë·ªông

result = simulate_fraction_experiment(fraction_matrix, initial_state, iterations)
print("Tr·∫°ng th√°i cu·ªëi c√πng c·ªßa h·ªá th·ªëng l√†:")
print(result)
```

### B√†i t·∫≠p 3.3 M√¥ ph·ªèng th√≠ nghi·ªám ·ªü h√¨nh 3.1 nh∆∞ng v·ªõi nhi·ªÅu khe h∆°n v·ªõi ùëõ l√† s·ªë khe v√† ùëö l√† s·ªë bia ƒë·∫°n. Ng∆∞·ªùi d√πng nh·∫≠p v√†o c√°c tham s·ªë tr√™n v√† x√°c su·∫•t m·ªói vi√™n ƒë·∫°n ƒëi t·ª´ khe ƒë·∫øn bia. Ch∆∞∆°ng tr√¨nh tr·∫£ v·ªÅ ma tr·∫≠n v√† vector k·∫øt qu·∫£.

```python
import numpy as np
from fractions import Fraction

def simulate_experiment(prob_matrix, initial_state, iterations):
    # Chuy·ªÉn ma tr·∫≠n v√† tr·∫°ng th√°i kh·ªüi t·∫°o th√†nh m·∫£ng numpy v·ªõi ki·ªÉu Fraction
    prob_matrix = np.array(prob_matrix, dtype=object)
    state = np.array(initial_state, dtype=object).reshape(-1, 1)
    
    # L·∫∑p l·∫°i qu√° tr√¨nh c·∫≠p nh·∫≠t tr·∫°ng th√°i n l·∫ßn
    for _ in range(iterations):
        state = np.dot(prob_matrix, state)
    
    return state

# ƒê·ªãnh nghƒ©a ma tr·∫≠n x√°c su·∫•t v√† tr·∫°ng th√°i kh·ªüi t·∫°o
prob_matrix = [
    [Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0)],
    [Fraction(1, 2), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0)], 
    [Fraction(1, 2), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0)], 
    [Fraction(0), Fraction(1, 3), Fraction(0), Fraction(1), Fraction(0), Fraction(0), Fraction(0), Fraction(0)], 
    [Fraction(0), Fraction(1, 3), Fraction(0), Fraction(0), Fraction(1), Fraction(0), Fraction(0), Fraction(0)], 
    [Fraction(0), Fraction(1, 3), Fraction(1, 3), Fraction(0), Fraction(0), Fraction(1), Fraction(0), Fraction(0)], 
    [Fraction(0), Fraction(0), Fraction(1, 3), Fraction(0), Fraction(0), Fraction(0), Fraction(1), Fraction(0)], 
    [Fraction(0), Fraction(0), Fraction(1, 3), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(1)], 
]

initial_state = [
    Fraction(1), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(0)
]
iterations = 2  # S·ªë l·∫ßn ho·∫°t ƒë·ªông

result = simulate_experiment(prob_matrix, initial_state, iterations)
print("Tr·∫°ng th√°i cu·ªëi c√πng c·ªßa h·ªá th·ªëng l√†:")
print(result)
```

### B√†i t·∫≠p 3.4 N√¢ng c·∫•p b√†i t·∫≠p 3.2 v·ªõi y√™u c·∫ßu ph·∫ßn t·ª≠ c·ªßa ma tr·∫≠n c√≥ th·ªÉ l√† s·ªë ph·ª©c

```python
import numpy as np
def simulate_complex_experiment(matrix, initial_state, n):
    # Chuy·ªÉn ma tr·∫≠n v√† tr·∫°ng th√°i kh·ªüi t·∫°o th√†nh m·∫£ng numpy v·ªõi ki·ªÉu complex
    matrix = np.array(matrix, dtype=complex)
    state = np.array(initial_state, dtype=complex).reshape(-1, 1)
    
    # L·∫∑p l·∫°i qu√° tr√¨nh c·∫≠p nh·∫≠t tr·∫°ng th√°i n l·∫ßn
    for _ in range(n):
        state = np.dot(matrix, state)
    
    return state

# ƒê·∫ßu v√†o c·ª• th·ªÉ c·ªßa b√†i to√°n v·ªõi ph·∫ßn t·ª≠ l√† s·ªë ph·ª©c
complex_matrix = [
    [1/np.sqrt(2) + 0j, 1/np.sqrt(2) + 0j, 0 + 0j],
    [0 - 1j/np.sqrt(2), 0 + 1j/np.sqrt(2), 0 + 0j],
    [0 + 0j, 0 + 0j, 0 + 1j]
]

initial_state = [
    1/np.sqrt(3) + 0j,
    0 + 2j/np.sqrt(15),
    0 + np.sqrt(2/5)*1j
]
iterations = 1  # S·ªë l·∫ßn ho·∫°t ƒë·ªông

result = simulate_complex_experiment(complex_matrix, initial_state, iterations)
print("Tr·∫°ng th√°i cu·ªëi c√πng c·ªßa h·ªá th·ªëng l√†:")
print(result)
```
